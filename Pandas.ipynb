数据合并

pandas.merge根据一个或多个键将多个DataFrame连接起来，类似数据库连接

pandas.concat可以沿着一个轴将多个对象堆叠起来

combine_first可以将重复数据编制在一起，用以填充另一个对象的缺失值

```python
dfn = pd.merge(dfa,dfb)#类似数据库，将两张表合并起来，卖家列是两张表都有的列，默认是使用相同列名，相当于内连接select * from a,b where a. = b.
dfn
pd.merge(dfa,dfb,on='卖家')#也可以指定使用列名合并
```

```python
pd.merge(dfa,dfb,how = 'outer')#可以指定外联 类似数据库的join外联方式，使用outer是默认左右互联
```

```python
pd.merge(dfa,dfb,how = 'left')#可以指定外联 类似数据库的join外联方式，使用left是指定左联
```

```python
pd.merge(dfa,dfb,how = 'right',on = '卖家')#可以指定外联 类似数据库的join外联方式，使用right是指定右联
```

```python
df1 = df[:3][['位置','卖家']]
df2 = df[:3][['价格','成交量']]
pd.merge(df1,df2,left_index=True,right_index=True)#将索引作为连接键
```

检测和过滤异常值

发现异常值和极端值的方法是对数据进行描述性统计，而使用describe函数则可以生成描述统计结果。

基于统计与数据分析
假设数据集满足正态分布，这里分布的平均值u，方差是a^2，如果x的值大于3a或者小于-3a，那么都可以认为是异常值

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
fig ,ax = plt.subplots(1,1,figsize=(8,5))
df = pd.read_csv("C:\\Users\\jacky_yin\\AppData\\Roaming\\data\\pandas data\\hz_weather.csv")
ax.hist(df['最低气温'],bins=20)
d = df["最低气温"]
#print(d.mean())
zscore = (d-d.mean())/d.std()
df['isOutlier'] = zscore.abs()>3
df['isOutlier'].value_counts()
```

箱形图分析，箱形图可以用来观察数据的整体的分布情况，利用中位数、25%分位数、75%分位数，上边界和下边界等统计量来描述数据的整体分布情况。

通过计算这些统计量，生成一个箱形图，箱体中的数据被认为是正常的，而在箱体上下边界之外的数据则被认为是异常

# 数据类型转换

在python中区分文本(string)、整形(Int)等数据类型，不同数据类型之间可以转换。

```python
import pandas as pd
df_pop = pd.read_csv("C:\\Users\\jacky_yin\\AppData\\Roaming\\data\\pandas data\\european_cities.csv")
df_pop.head()
```

```python
type(df_pop.Population[0])
```

```python
df_pop["NumericPopulation"]=df_pop.Population.apply(lambda x:int(x.replace(",","")))
#lambda表示下面的内容是一个函数，对每一行用apply操作，去掉population每个元素的逗号并转为integer
#apply很方便，不用循环
df_pop.head()
```

```python
df_pop["State"].values[:3]#取state列中的前三个数据
```

```python
df_pop["State"] = df_pop["State"].apply(lambda x:x.strip())
df_pop["State"].values[:3]
```
